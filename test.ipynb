```python
def generate_image(pipe, prompt, params):
  img = pipe(prompt, **params).images

  num_images = len(img)
  if num_images>1:
    fig, ax = plt.subplots(nrows=1, ncols=num_images)
    for i in range(num_images):
      ax[i].imshow(img[i]);
      ax[i].axis('off');

  else:
    fig = plt.figure()
    plt.imshow(img[0]);
    plt.axis('off');
  plt.tight_layout()
```

```python
prompt = "HORSE"

params = {}
```

```python
generate_image(pipe, prompt, params)
```

```python
#height width
params = {"height": 1024, "width": 1024}

generate_image(pipe, prompt, params)
```
